```python
import pandas as pd
```

```python
#After loading the dataset I displayed the first 60 rows to see any Nan values
data = pd.read_excel('/content/Detailed_LA_20232024.xlsx', sheet_name='A8',header=4,)
data.head(60)
```

```python
#This code helps me visualise the datatypes, unnamed colums that may need to be named/removed and coulmns with missing values
data.info()
```

```python
#With this i named a needed unanmed coulmn and "Total owed a prevention or relief duty1" in order to not have them removed with the method below that dropps columns that are unnamed and colums that have the word total in(data = data.loc[:, ~data.columns.str.contains(r'^(Unnamed|Total)', case=False)])
data = data.rename(columns={})
data = data.rename(columns={
    "Unnamed: 1": "Cities",
    "Total owed a prevention or relief duty1": "Sum of owed a prevention or relief duty",
    "Not known2": "Not known"
})

#With this it displays if my renaming was correctly executed
print(data.head())

#This Drop columns with less than 2 non-null values
data = data.dropna(axis=1, thresh=2)

#This removes rows containing '..' in any column
mask = data.applymap(lambda x: '..' in str(x)).any(axis=1)
data = data[~mask]

#This drops rows that are entirely empty
data = data.dropna(axis=0, how='all')

#This drops all unnamed columns and those containing 'Total'
data = data.loc[:, ~data.columns.str.contains(r'^(Unnamed|Total)', case=False)]

data = data.iloc[:291, :]  #This Keeps only rows 0 to 291 (total 292 rows) due to irrelevent text due from referencing that came with uncleaned dataset

#This displays the first 50 rows
print(data.head(50))

#Once i have cleaned the data i saved it to cleaned-A8-data.csv which will be split to 2 other datasets that will display the regions and the cities of the UK
data.to_csv('cleaned-A8-data.csv',index=False)
```
