```python
data=pd.read_csv('/content/cleaned-A10-data.csv')
#London duplicate
data=data.drop(index=data.index[1])
cities=data[['Cities']]
cities_df = pd.read_csv("/content/worldcities.csv")
filtered_df = cities_df[(cities_df['country'] == 'United Kingdom') & (cities_df['city'].isin(cities['Cities']))]
filtered_df = filtered_df.drop(columns=['city_ascii','country', 'iso2', 'iso3', 'admin_name', 'capital', 'population', 'id'])
filtered_df = filtered_df.reset_index(drop=True)
print(filtered_df.shape)
pd.set_option('display.max_rows', 200)
filtered_df.head(100)
```

```python
data = data[data['Cities'].isin(filtered_df['city'])]
print(data.shape)
data.head(100)
```

```python
#Merge and add the 'total_population' column where cities match
a10 = pd.merge(
    filtered_df,
    data[['Cities','Sum of owed a prevention or relief duty','Student / training']],  # Select only the needed column
    left_on='city',
    right_on='Cities',
    how='inner'  # Keep only cities present in BOTH datasets
)
```

```python
#to drop cities column cause its redundant
a10 = a10.drop(columns=['Cities'])
a10.head(115)
```

```python
print(a10.shape)
print(a10A.shape)
```

```python
# prompt: check if there are some names in city in a8 that is not in a10

# Assuming 'city' column exists in both DataFrames
a10_cities = set(a10['city'].dropna())  # Use dropna to handle potential missing values
a10a_cities = set(a10A['city'].dropna())

cities_not_in_a10a = a10_cities - a10a_cities

if cities_not_in_a10a:
    print("Cities in a10 not present in a10a:")
    for city in cities_not_in_a10a:
        print(city)
else:
    print("All cities in a8 are present in a10a.")
```

```python
#we want to add this 5 rows to my a10a dataset

import pandas as pd
# Assuming 'city' column exists in both DataFrames and you want to add rows from a10 to a10a
# based on matching cities.
new_rows = a10[(a10['city']=='Durham')|(a10['city']=='Bristol')|(a10['city']=='London')|(a10['city']=='Lincoln')|(a10['city']=='Newcastle under Lyme')]

# Concatenate the new rows to a10a and ***Make sure we are not running this twice***
merged_df = pd.concat([a10A, new_rows], ignore_index=True)

print(merged_df.shape)
merged_df.tail(10) #Inspect last few rows to confirm addition
```

#start the clustering

```python
#datanew=data[['city','Sum of owed a prevention or relief duty','Student / training']]
```

```python
from sklearn.cluster import KMeans
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(merged_df[['Sum of owed a prevention or relief duty','Student / training']])  # Use X_scaled if standardized
    inertias.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 11), inertias, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
merged_df['cluster']=kmeans.fit_predict(merged_df[['Sum of owed a prevention or relief duty','Student / training']])
merged_df['cluster']=merged_df['cluster'].astype('category')
print(merged_df.describe(include='all'))
merged_df.head(15)
```
